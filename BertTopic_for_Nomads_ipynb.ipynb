```python
import numpy as np
import pandas as pd
docs = pd.read_csv("dncommentswsentiment.csv")

#docs = docs.dropna(subset=['TSR'])
reviews = docs['comment']
reviews = reviews.dropna()
```

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import sent_tokenize
review_blob = " ".join(reviews)
reviewsents = sent_tokenize(review_blob)
```

```python
nltk.download('stopwords')
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

tokens = nltk.word_tokenize(review_blob)

filtered_tokens = [w.lower() for w in tokens if not w.lower() in stop_words]
filtered_tokens = [w for w in filtered_tokens if w.isalnum()]

#Create your bigrams
bgs = nltk.bigrams(filtered_tokens)


#compute frequency distribution for all the bigrams in the text
fdist = nltk.FreqDist(bgs)
mincount = 10
for k,v in fdist.items():
    if v> mincount:
      print (k,v)
```

```python
wdist = nltk.FreqDist(filtered_tokens)
mincount = 200
for k,v in wdist.items():
    if v> mincount:
      print (k,v)
```

```python
from bertopic import BERTopic
#from sklearn.datasets import fetch_20newsgroups

#docs = fetch_20newsgroups(subset='all',  remove=('headers', 'footers', 'quotes'))["data"]

#docs = pd.read_excel("Liminal July 20 All coded.xlsx")

#reviews = docs['Review']
from bertopic.representation import KeyBERTInspired

# Fine-tune your topic representations
representation_model = KeyBERTInspired()
#topic_model = BERTopic(representation_model=representation_model)
seed_topic_list = [["money","well being","safety","gear","politics"]]

topic_modelg = BERTopic(representation_model=representation_model,seed_topic_list=seed_topic_list,n_gram_range=(1,2)) #,seed_topic_list=seed_topic_list) #n_gram_range=(2,2)
topics, probs = topic_modelg.fit_transform(reviews)
```
